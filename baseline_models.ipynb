```python
keras_mod = keras.Sequential([
    keras.layers.Input(shape=(67,)),  # 67 features
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(32, activation='relu'),
    # 3 class
    keras.layers.Dense(3, activation='softmax')]
)

keras_mod.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

models = [
    # sgd
    sklearn.linear_model.SGDClassifier(loss='log_loss', max_iter=1000, tol=1e-3),
    # naive bayes
    sklearn.naive_bayes.MultinomialNB(),
    # keras
    keras_mod
]
```

```python
for model in models:
    modeler = model_wrapper.Modeler(model, trainer_type='keras' if isinstance(model, keras.Sequential) else 'sklearn')
    modeler.training_loop(num_epochs=100, data_type='simple')
```

```python
# for each model: evaluate
results = {}

for model in models:
    modeler = model_wrapper.Modeler(model, trainer_type='keras' if isinstance(model, keras.Sequential) else 'sklearn')
    results[model] = modeler.evaluate(data_type='simple')
```

```python
# print results and show confusion matrix for first model
results_sgd = results[models[0]]
print("SGD Classifier Results:")
print("Accuracy:", results_sgd['accuracy'])
print("Precision:", results_sgd['precision'])
print("Recall:", results_sgd['recall'])
print("F1 Score:", results_sgd['f1_score'])

from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt

disp = sklearn.metrics.ConfusionMatrixDisplay(confusion_matrix=results_sgd['confusion_matrix'], display_labels=[0, 1, 2])
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix for SGD Classifier")
plt.show()
```

```python
# show confusion matrix for second model
results_nb = results[models[1]]
print("Naive Bayes Results:")
print("Accuracy:", results_nb['accuracy'])
print("Precision:", results_nb['precision'])
print("Recall:", results_nb['recall'])
print("F1 Score:", results_nb['f1_score'])

disp = sklearn.metrics.ConfusionMatrixDisplay(confusion_matrix=results_nb['confusion_matrix'], display_labels=[0, 1, 2])
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix for Naive Bayes")
plt.show()
```

```python
# show confusion matrix for third model
results_keras = results[models[2]]
print("Keras Sequential Model Results:")
print("Accuracy:", results_keras['accuracy'])
print("Precision:", results_keras['precision'])
print("Recall:", results_keras['recall'])
print("F1 Score:", results_keras['f1_score'])

disp = sklearn.metrics.ConfusionMatrixDisplay(confusion_matrix=results_keras['confusion_matrix'], display_labels=[0, 1, 2])
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix for Keras Sequential Model")
plt.show()
```
